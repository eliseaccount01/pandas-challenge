```python
#import dependencies 
import pandas as pd
```

```python
#set path and read to csv
data_path ='resources/purchase_data.csv'
heroes_data=pd.read_csv(data_path)

heroes_data.head()
```

```python
#find total players
total_players = len(heroes_data["SN"].unique())
total_players
```

```python
#create purchase analysis dataframe 
unique_names = heroes_data["Item Name"].nunique()
avg_price = round(heroes_data["Price"].sum()/heroes_data["Price"].count(),2)
total_purchases = heroes_data["Purchase ID"].count()
total_revenue = heroes_data["Price"].sum()
purchase_analysis = pd.DataFrame({"Number of Unique Items":[unique_names],"Average Purchase Price":[avg_price],
                    "Total Number of Purchases":[total_purchases],"Total Revenue":[total_revenue]})
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:,.2f}".format)
purchase_analysis
```

```python
#drop all duplicate SN to obtain gender demographics
demo_total = heroes_data.drop_duplicates(subset ='SN',keep = 'first',
                                        inplace = False)
gender_total = demo_total["Gender"].value_counts()
percent = round(gender_total/total_players *100,2).astype(float).map("{:,.2f}%".format)

demographics = pd.DataFrame({"Total Number":gender_total,"Pecentage":percent})
demographics
```

```python
# Purchasing Analysis (Gender)
total=heroes_data.groupby(["Gender"]).sum()["Price"]
count =heroes_data.groupby(["Gender"]).count()["Purchase ID"]
avg_cost = heroes_data.groupby(["Gender"]).mean()["Price"]
avg_cost_per = round(total/gender_total,2)

purchase_demo_df = pd.DataFrame({ "Purchase Value":count,
                            "Average Purchase Price":avg_cost.astype(float).map("${:,.2f}".format),
                            "Total Purchase Count": total.astype(float).map("${:,.2f}".format), 
                           "Avg Total Purchase per Person":avg_cost_per.astype(float).map("${:,.2f}".format)})
purchase_demo_df
```

```python
# Age Demographics
bins =[0,9,14,19,24,29,34,39,100] 
labels = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
demo_total["Ages"]=pd.cut(demo_total["Age"],bins,labels=labels)

total_count = demo_total.groupby(["Ages"]).count()["Age"]
percent_players = round(total_count/total_players*100,2).astype(float).map("{:,.2f}%".format)
age_demographics = pd.DataFrame({"Total Count": total_count, "Percent of Players":percent_players})

age_demographics
```

```python
#Purchasing Analysis (Age) 

bins =[0,9,14,19,24,29,34,39,100] 
labels = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
heroes_data["Ages"]=pd.cut(heroes_data["Age"],bins,labels=labels)


total_purchase=heroes_data.groupby("Ages").sum()["Price"]
purchase_count = heroes_data.groupby("Ages").count()["Purchase ID"]
avg_purchase =(total_purchase/purchase_count).astype(float).map("${:,.2f}".format)
total_count = demo_total.groupby(["Ages"]).count()["Age"]
avg_per_person = (total_purchase/total_count).astype(float).map("${:,.2f}".format)

Age_Purchasing_df = pd.DataFrame({"Purchase Count": purchase_count ,"Average Purchase Price": avg_purchase,
                                  "Total Purchase Value":total_purchase.astype(float).map("${:,.2f}".format), "Avg Total Purchase per Person": avg_per_person})
Age_Purchasing_df
```

```python
#Top Spenders
total_value_purchase = heroes_data.groupby("SN").sum()["Price"]
purchase_count = heroes_data.groupby("SN").count()["Purchase ID"]
avg_purchase_count = (total_value_purchase/purchase_count).astype(float).map("${:,.2f}".format)
top_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_count,
           "Total Purchase Value": total_value_purchase.astype(float).map("${:,.2f}".format)})
top_five = top_df.sort_values("Total Purchase Value", ascending = False)
top_five.head()
```

```python
#Most popular items 
counts=heroes_data.groupby(["Item ID","Item Name"]).count()["Price"]
sum=heroes_data.groupby(["Item ID","Item Name"]).sum()["Price"].astype(float).map("${:,.2f}".format)
price = heroes_data.groupby(["Item ID","Item Name"]).mean()["Price"].astype(float).map("${:,.2f}".format)

most_poopular = pd.DataFrame({"Purchase Count":counts, "Item Price": price, "Total Purchase Value": sum})
most_poopular.sort_values("Purchase Count", ascending = False).head()
```
